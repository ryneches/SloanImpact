```python
%pylab inline
import requests
import json
```

```python
awards = []

url = 'http://api.nsf.gov/services/v1/awards.json'

offset = 0
while True :
    params = { 'keyword' : 'microbiology+built+environment',
               'offset'  : offset }
    response = json.loads( requests.get(url, params=params).content )
    if not response['response'].has_key( 'award' ) : break
    for award in response['response']['award'] :
        awards.append(award)
    print len(awards)
    offset += 1
```

```python
import pandas as pd
```

```python
nsf = pd.DataFrame.from_dict(awards)

nsf.date = pd.to_datetime( nsf['date'], infer_datetime_format=True)
```

```python
import seaborn

nsf.groupby(nsf.date.dt.year).count().plot(kind="bar")
```
